# Modeling Classification Notebook

## Introduction

In this notebook, we will build and evaluate several **classification models** to predict whether a stock will **outperform the S&P 500 by more than 2%**. The models we train will be based on the financial indicators we've cleaned, engineered, and processed in the previous steps. By comparing multiple machine learning algorithms, we aim to identify the most effective model for predicting stock performance.

### Project Goal

The ultimate goal of this project is to build a machine learning model that can select stocks with a high likelihood of outperforming the S&P 500 by more than 2%. By successfully predicting stock outperformance, we aim to create a portfolio that consistently beats the market.

### Key Models We Will Evaluate:
1. **Logistic Regression**: A simple, interpretable baseline model.
2. **Decision Trees**: A tree-based model that splits the data into smaller subsets to make predictions based on learned rules.
3. **Random Forest Classifier**: A decision-tree-based ensemble model that provides feature importance.
4. **Gradient Boosting Models (XGBoost, LightGBM)**: Advanced models designed to handle complex patterns and interactions.
5. **Support Vector Machines (SVM)**: A model that finds the optimal boundary between outperforming and underperforming stocks.
6. **K-Nearest Neighbors (KNN)**: A simple model that classifies based on the similarity to neighboring stocks.
7. **Artificial Neural Networks (Sequential ANN)**: A deep learning model that captures non-linear relationships in the data.

### Steps in This Notebook:
- **Model Training**: We will train each model on the dataset.
- **Model Evaluation**: Using metrics such as **accuracy**, **precision**, **recall**, **F1 score**, and **AUC-ROC**, we will evaluate each model’s performance.
- **Model Comparison**: We will compare the results from each model to identify which one provides the best prediction accuracy for stock outperformance.

Once the best-performing model is identified, we will proceed to **hyperparameter tuning** in a separate notebook to further optimize the model's performance.

### Loading Dataset

```python
df = pd.read_csv('data/for_modeling_datasets/full_Financial_Data_for_modeling.csv')
df.head()
```

```python
# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(message)s')

def preprocess_data(full_df, target_col='Outperform', drop_cols=['PRICE VAR [%]', 'Class', 'Symbol']):
    """
    Preprocess the dataset by splitting into training, validation, and test sets,
    balancing classes with SMOTE, and standardizing features with StandardScaler.
    
    Parameters:
    - full_df: pd.DataFrame, the full merged dataset
    - target_col: str, the target variable name ('Outperform')
    - drop_cols: list, columns to drop due to data leakage or irrelevance
    
    Returns:
    - X_train, X_val, X_test: feature matrices for train, validation, and test sets
    - y_train, y_val, y_test: target variables for train, validation, and test sets
    - scaler: the StandardScaler object fitted on the training data
    """
    
    # Split the data by year
    train_data = full_df[full_df['Year'].isin([2014, 2015, 2016])]
    val_data_first_half = full_df[(full_df['Year'] == 2017) & (full_df.index < len(full_df) / 2)]
    val_data_second_half = full_df[(full_df['Year'] == 2017) & (full_df.index >= len(full_df) / 2)]
    test_data = full_df[full_df['Year'] == 2018]
    
    # Combine training and validation data for first half of 2017
    train_data = pd.concat([train_data, val_data_first_half], axis=0)
    
    logging.info(f"Training data size: {train_data.shape}")
    logging.info(f"Validation data size (second half 2017): {val_data_second_half.shape}")
    logging.info(f"Test data size (2018): {test_data.shape}")
    
    # Drop the columns that may cause data leakage
    X_train = train_data.drop(columns=drop_cols + [target_col])
    y_train = train_data[target_col]
    
    X_val = val_data_second_half.drop(columns=drop_cols + [target_col])
    y_val = val_data_second_half[target_col]
    
    X_test = test_data.drop(columns=drop_cols + [target_col])
    y_test = test_data[target_col]
    
    # Handle class imbalance using SMOTE on the training data
    smote = SMOTE(random_state=42)
    X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
    
    logging.info(f"After SMOTE, training data size: {X_train_smote.shape}")
    
    # Standardize the features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_smote)
    X_val_scaled = scaler.transform(X_val)
    X_test_scaled = scaler.transform(X_test)
    
    return X_train, X_val, X_test, y_train, y_val, y_test

def train_and_evaluate_model(model, X_train, y_train, X_val, y_val):
    """
    Train the model and evaluate it on validation data, including confusion matrix and classification report.
    
    Parameters:
    - model: the machine learning model to train
    - X_train: training features
    - y_train: training labels
    - X_val: validation features
    - y_val: validation labels
    
    Returns:
    - classification report and confusion matrix
    """
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on validation data
    y_pred = model.predict(X_val)
    
    # Confusion Matrix
    cm = confusion_matrix(y_val, y_pred)
    ConfusionMatrixDisplay(cm).plot()
    plt.title(f"Confusion Matrix for {model.__class__.__name__}")
    plt.show()
    
    # Classification Report
    report = classification_report(y_val, y_pred)
    print(f"Classification Report for {model.__class__.__name__}:\n")
    print(report)
    
    return y_pred

# Run
X_train, X_val, X_test, y_train, y_val, y_test = preprocess_data(df)
```

### 1. Logistic Regression: Model Evaluation

```python
# Initialize Logistic Regression model
log_reg = LogisticRegression(random_state=42)

# Call the function to train and evaluate the model
y_pred_log_reg = train_and_evaluate_model(log_reg, X_train, y_train, X_val, y_val)
```

The **Logistic Regression** model shows the following results on the validation set:

- **Accuracy**: 60%
- **Class 0 (Non-Outperformers)**:
  - **Precision**: 0.67
  - **Recall**: 0.77
  - **F1-Score**: 0.71
- **Class 1 (Outperformers)**:
  - **Precision**: 0.37
  - **Recall**: 0.26
  - **F1-Score**: 0.31

#### Key Insights:
- The model performs reasonably well at identifying **non-outperforming stocks** (Class 0) with a recall of 0.77.
- However, it struggles with **outperforming stocks** (Class 1), with a low recall of 0.26 and an F1-score of 0.31.
- **Class imbalance** might be impacting the model's ability to correctly identify outperformers, leading to a weighted average F1-score of 0.58.

This model may benefit from hyperparameter tuning or the use of more sophisticated models to better detect stocks that outperform the S&P 500.

### Artificial Neural Network (ANN): Model Evaluation

```python
# Initialize the Sequential Neural Network model
ann = Sequential()

# Add layers to the ANN
ann.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
ann.add(Dense(32, activation='relu'))
ann.add(Dense(1, activation='sigmoid'))

# Compile the model
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
ann.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)

# Evaluate the model on validation data
y_pred_ann = (ann.predict(X_val) > 0.5).astype("int32")

# Confusion Matrix
cm = confusion_matrix(y_val, y_pred_ann)
ConfusionMatrixDisplay(cm).plot()
plt.title(f"Confusion Matrix for ANN")
plt.show()

# Classification Report
report = classification_report(y_val, y_pred_ann)
print(f"Classification Report for ANN:\n")
print(report)
```

The **ANN** model shows the following results on the validation set:

- **Accuracy**: 66%
- **Class 0 (Non-Outperformers)**:
  - **Precision**: 0.66
  - **Recall**: 1.00
  - **F1-Score**: 0.79
- **Class 1 (Outperformers)**:
  - **Precision**: 0.43
  - **Recall**: 0.00
  - **F1-Score**: 0.01

#### Key Insights:
- The model performs exceptionally well for **non-outperformers** (Class 0) with a perfect recall of 1.00, meaning that almost all non-outperforming stocks are correctly classified.
- However, it fails to identify **outperformers** (Class 1) with a **recall of 0.00**, meaning it classifies almost all the actual outperformers as non-outperformers.

This model is highly biased towards Class 0, making it unsuitable for identifying outperforming stocks in its current state. Further adjustments or a different architecture may be needed to improve its performance on outperformers.

# Model Comparison and Conclusion

Our goal is to build a model that can **beat the market** by identifying stocks that will **outperform the S&P 500**. However, it is equally important to **avoid losing money**, meaning the model should avoid predicting underperforming stocks as outperforming ones.

Let's compare the models based on **accuracy**, **precision**, **recall**, and **F1-score** for both classes (Outperformers and Non-Outperformers):

#### Key Takeaways:

1. **Best Model for Outperformers**:
   - **Logistic Regression** and **XGBoost** have the best balance in terms of identifying **outperformers** (Class 1), with relatively better F1-scores (0.31 and 0.33, respectively) compared to other models. 
   - **K-Nearest Neighbors** and **Decision Trees** show moderate performance in identifying outperformers but are not exceptional.
   - **ANN** fails to detect outperformers entirely (Recall = 0.00), making it the least useful model in this context.

2. **Best Model for Non-Outperformers**:
   - **SVC** and **Random Forest** perform the best at identifying **non-outperformers** (Class 0) with F1-scores of **0.77** and **0.76**, respectively.
   - **ANN** is extremely biased towards non-outperformers, perfectly identifying them with a recall of **1.00**, but at the cost of ignoring outperformers.

3. **Balanced Models**:
   - **XGBoost** offers the best balance between **precision** and **recall** for both classes, suggesting it could be a good candidate for further tuning.
   - **Logistic Regression** is also relatively balanced, though it doesn't outperform XGBoost.

4. **Model for Risk Mitigation**:
   - Given our goal to avoid **losing money**, we should prioritize models that **minimize false positives** for Class 1 (i.e., predicting a stock as outperforming when it is not).
   - **XGBoost** and **Logistic Regression** are the safest options, as they maintain a reasonable precision for Class 1 (0.40 and 0.37, respectively), meaning fewer incorrect predictions of outperforming stocks.

#### Final Conclusion:

- **XGBoost** appears to be the most promising model, providing a decent balance between **recall** and **precision** for both classes. It minimizes the risk of falsely predicting outperforming stocks and provides a reasonable overall F1-score for outperformers (0.33).
- **Logistic Regression** is a simpler model but still offers reasonable performance and could be a good alternative if we prioritize model interpretability.
- **ANN** and **SVC** are not suitable for this use case, as they fail to adequately capture the complexity of identifying outperformers.
- Further improvements could come from **hyperparameter tuning** of the **XGBoost** model or experimenting with ensemble methods that combine the strengths of multiple models (e.g., stacking).

Next steps should include focusing on improving **XGBoost** through hyperparameter tuning and additional techniques to handle the **class imbalance** in the dataset.
